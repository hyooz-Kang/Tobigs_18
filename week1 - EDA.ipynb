관심있게 살펴볼 변수에 대한 분포를 확인해본 결과 특별한 이상치, 혹은 결측치는 없는 것으로 결론 내릴 수 있었다.
유의미한 시각화를 위해 가설을 세워보자.


*   타겟값인 퇴직 여부는 Age와 상관관계가 있을 것이다.
*   업무 분야에 따라 퇴직 여부 정도가 다를 것이다
* 업무환경 / 업무종류에 관한 만족도에 따라 퇴직 여부 정도가 다를 것이다.
* 일한 회사 수에 따라 퇴직 정도가 다를 것이다.
* 회사와 집 간의 거리에 따라 퇴직 정도가 다를 것이다.


```python
#Age와 퇴직 여부에 관해 시각화 진행
age_df = df[["Age", "Attrition"]]
```

```python
age_y = age_df[age_df["Attrition"] == "Yes"]
```

```python
age_n = age_df[age_df["Attrition"] == "No"]
```

```python
plt.hist(age_y["Age"], label='yes', color='red', alpha=0.2, bins=50, density=True) #퇴직한 직원
plt.hist(age_n["Age"], label='no', color='blue', alpha=0.2, bins=50, density=True) #퇴직하지 않은 직원
plt.hist(df["Age"], label='total',color='green', alpha=0.2, bins=50, density=True) #토탈
plt.legend()
plt.show()
```

퇴직하지않은(no) 직원에 대한 히스토그램과 전체 히스토그램의 분포는 비슷한 반면,

*   퇴직한 직원(yes)의 분포는 젊은층(18 ~ 30대 초반)에서 크게 나타나고
*   중장년층(30대 중반 ~ 50)에서 적게 나타나고
* 노년층(50대 이상~)에서 다시 조금 크게 나타나는 것을 알 수 있다.

-> 젊은층의 퇴직 정도가 크고, 중장년층은 퇴직 정도가 적다는 결론 가능



```python
# 업무 분야에 따른 퇴직 정도 시각화
depart_df = df[["Department", "Attrition", "EmployeeNumber"]].groupby(["Department", "Attrition"]).count().unstack()
```

*   Sales Representative, Laboratory Technician, Sales Executive, Research Scientist 분야에 종사하는 직원들의 경우 퇴직 비율이 높다.

```python
#업무환경의 만족도에 따른 퇴직 정도 시각화
ES_df = df[["EnvironmentSatisfaction", "Attrition", "EmployeeNumber"]].groupby(["EnvironmentSatisfaction", "Attrition"]).count().unstack()
```

만족도를 전체적으로 파악할 수 있는 새로운 파생변수를 생성하여, 그에 따른 퇴직 비율을 파악해보자.

```python
#업무 환경, 업무 종류, 대인관계 만족도를 총체적으로 파악할 수 있는 "만족도"라는 새로운 파생변수를 생성해보자.
df["satisfaction"] = df["JobSatisfaction"] + df["EnvironmentSatisfaction"] + df["RelationshipSatisfaction"]
```

```python
plt.hist(df["satisfaction"])
```

```python
Satisfaction_df = df[["satisfaction", "Attrition", "EmployeeNumber"]].groupby(["satisfaction", "Attrition"]).count().unstack()
```

```python
Satisfaction_df.plot.bar(stacked=True)
```

업무 환경, 업무 종류, 대인관계 만족도를 모두 포함하는 "만족도"에 따라 퇴직 비율이 다른 것을 확인할 수 있다. 특히 전체 만족도가 3,4인 경우 많은 수가 퇴직한 것을 알 수 있다.
